```python
import pandas as pd
```

```python
hors_carte = pd.read_excel("data/invaders_todo_hors_carte.xlsx")
hors_carte
```

```python
remybd = pd.read_excel("data/remybd_clean_data.xlsx")
remybd = remybd.rename(columns={"localisation":"name"})
remybd
```

```python
pd.merge(remybd, hors_carte,on="code")
```

```python
# Merge df1 with df2 on 'code' with suffixes to distinguish the Latitude columns
merged_df = hors_carte.merge(remybd, on='code', suffixes=('', '_df2'),how="left")

# Impute missing Latitude in hors_carte with Latitude from df2
merged_df['Latitude'] = merged_df['Latitude'].combine_first(merged_df['Latitude_df2'])
merged_df['Longitude'] = merged_df['Longitude'].combine_first(merged_df['Longitude_df2'])
merged_df['name'] = merged_df['name'].combine_first(merged_df['name_df2'])

# Drop the extra Latitude_df2 column
hors_carte_imputed = merged_df.drop(columns=['Latitude_df2','Longitude_df2','name_df2'])
hors_carte_imputed
```
