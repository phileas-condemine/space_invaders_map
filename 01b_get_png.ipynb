```python
import pandas as pd
data = pd.read_excel("data/20241111_export_invader_spotter_art_listing.xlsx")
data = data[[0]]
data = data.rename(columns={0:"relative_url_raw"})
data['relative_url'] = data['relative_url_raw'].str.extract(r'<img[^>]*src="([^"]+)"', expand=False)
data['url'] = "https://www.invader-spotter.art/" + data['relative_url']
data
```

```python
import os
import requests

# Function to download and save a PNG file
def download_png(url, save_dir):
    try:
        # Get the PNG name from the URL
        png_name = os.path.basename(url)
        # Full path to save the PNG file
        save_path = os.path.join(save_dir, png_name)

        # Send a GET request to the URL
        response = requests.get(url)
        response.raise_for_status()  # Check if the request was successful

        # Save the content to a file
        with open(save_path, "wb") as file:
            file.write(response.content)
        
        print(f"Downloaded and saved {png_name}")
    except requests.exceptions.RequestException as e:
        print(f"Failed to download {url}: {e}")
```

```python


# List of URLs to download
urls = data['url'].tolist()

# Directory to save the images
save_dir = "img/"

# Create the directory if it doesn't exist
os.makedirs(save_dir, exist_ok=True)


# Loop over the list of URLs and download each PNG file
for url in urls:
    download_png(url, save_dir)
```
