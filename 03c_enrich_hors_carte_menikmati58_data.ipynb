```python
def standardize_code(code):
    # Handle cases where the input is not a string
    if not isinstance(code, str):
        return code
    # Define a function to standardize the string format
    match = re.match(r"PA.{1}(\d{2,4})", code)
    if match:
        # Extract the number part and pad with zeros to make it 4 digits
        number_part = match.group(1)
        standardized_code = f"PA_{int(number_part):04d}"
        return standardized_code
    return code  # Return the original code if it doesn't match the pattern
```

```python
import re
data = pd.read_csv("data/20241208_umap_menikmati58_Laure.csv")
# data = remybd.rename(columns={"localisation":"name"})
data['name'] = data['name'].fillna("")
data['name'] = data['name'].str.replace("-","_")

data['code'] = data['name'].apply(lambda x: re.findall(r'PA.{1}[0-9]{2,4}', x))
data = data.explode('code')
data['code'] = data['code'].str.replace("-","_")
data['code'] = data['code'].apply(standardize_code)
data = data[data['code'].str.startswith(pat="PA_",na=False)]

data['localisation'] = data['name'].str.replace(pat=r'PA.{1}[0-9]{2,4}',repl="",regex=True)
data['localisation'] = data['localisation'].str.replace(pat=r"^[ ^:&-]*",repl="",regex=True)
data = data.sort_values("code")
data
```

```python
useful = pd.merge(data, hors_carte[hors_carte['Latitude'].isna()][['code']],on="code")
print(useful['name'].unique())
print(useful['description'].unique())
```

```python
# Merge df1 with df2 on 'code' with suffixes to distinguish the Latitude columns
merged_df = hors_carte.merge(data[['code','Latitude','Longitude','description']], on='code', suffixes=('', '_df2'),how="left")

# Impute missing Latitude in hors_carte with Latitude from df2
merged_df['Latitude'] = merged_df['Latitude'].combine_first(merged_df['Latitude_df2'])
merged_df['Longitude'] = merged_df['Longitude'].combine_first(merged_df['Longitude_df2'])
merged_df['name'] = merged_df['name'].combine_first(merged_df['description'])

# Drop the extra Latitude_df2 column
hors_carte_imputed = merged_df.drop(columns=['Latitude_df2','Longitude_df2','description'])
hors_carte_imputed
```
