```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Specify the path to the web page
path = "https://www.invader-spotter.art/cherche.php"  # Replace with your actual URL

# Initialize the WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Open the web page
driver.get(path)

# Perform any additional actions on the web page
# For example, find an element by its name
# element = driver.find_element(By.NAME, "q")


```

```python
# Effectuer manuellement la recherche : cocher Paris + tous les arrondissements > recherche pour arriver sur la page de listing.php
```

```python

def extract_rows(driver):
    # Get the content of the first <table> tag
    table = driver.find_element(By.TAG_NAME, 'table')

    # Extract rows and columns
    rows = table.find_elements(By.CSS_SELECTOR, 'tr.haut')
    data = []
    for row in rows:
        cols = row.find_elements(By.TAG_NAME, 'td')
        row_data = []
        for col in cols:
            # Split the text by <br> tags
            cell_data = col.get_attribute('innerHTML').split('<br>')
            cell_data = [field.strip() for field in cell_data if field.strip()]  # Strip whitespace and remove empty fields
            row_data.extend(cell_data)
        data.append(row_data)

    # Convert to Pandas DataFrame
    df = pd.DataFrame(data)
    return(df)
```

```python
res = []
page_data = extract_rows(driver)
res.append(page_data)
res
```

```python

# Create an empty DataFrame to store all the data
# Loop through pages 2 to 40
for page in range(2, 32):
    # Execute JavaScript to change the page
    driver.execute_script(f"javascript:changepage({page})")
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.TAG_NAME, 'table')))

    # Extract the rows and append to the all_data DataFrame
    page_data = extract_rows(driver)
    res.append(page_data)
```

```python
df = pd.concat(res,ignore_index=True)
df
```

```python
df.to_excel("data/20241111_export_invader_spotter_art_listing.xlsx")
```
