```python
def pad_number(my_str):
    parts = my_str.split('_')
    parts[1] = parts[1].zfill(4)
    return '_'.join(parts)
```

```python
import re

file_names = os.listdir("phone_data/")
pattern = r'([A-Z]+_+[0-9]+)'
flashed = [re.search(pattern, file_name).group(0) for file_name in file_names if re.search(pattern, file_name)]
flashed = [pad_number(file_name) for file_name in flashed]
flashed = list(set(flashed))

with open('data/flashed.txt', 'w') as file:
    for flash in flashed:
        file.write(flash + '\n')

```
